```python
#VAN DER POL AND ABS VAN DER POL MODELS

#VAN DER POL OSCILLATOR
#
#  X' = V
#  V' = M ( 1 - X^2 )V - X

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the Van der Pol oscillator equations
def van_der_pol(t, y, m):
    x, v = y
    dxdt = v
    dvdt = m * (1 - x**2) * v - x
    return [dxdt, dvdt]

# Simulation function
def simulate_van_der_pol(m, Tmin, Tmax):
    # Set random initial conditions in the rectangle [-0.1, 0.1] * [-0.1, 0.1]
    x0 = np.random.uniform(-0.1, 0.1)
    v0 = np.random.uniform(-0.1, 0.1)
    
    # Define the time span and evaluation points with a minimum of 30,000 points
    t_span = (0, Tmax)
    t_eval = np.linspace(Tmin, Tmax, 30000)  # Ensures 30,000 points in [Tmin, Tmax]
    
    # Solve the ODE using a stiff solver
    sol = solve_ivp(van_der_pol, t_span, [x0, v0], args=(m,), t_eval=t_eval, method='LSODA')
    
    # Plot results in the interval [Tmin, Tmax]
    plt.figure(figsize=(10, 5))
    plt.plot(sol.t, sol.y[0], label='x (position)')
    plt.plot(sol.t, sol.y[1], label='v (velocity)')
    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.title(f'Van der Pol Oscillator Simulation (m={m}, Tmin={Tmin}, Tmax={Tmax})')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Phase plot
    plt.figure(figsize=(6, 6))
    plt.plot(sol.y[0], sol.y[1], color='purple')
    plt.xlabel('x (position)')
    plt.ylabel('v (velocity)')
    plt.title('Phase Plot of Van der Pol Oscillator')
    plt.grid()
    plt.show()

```

```python
# User input for m, Tmin, and Tmax
m = float(input("Enter the parameter m: "))
Tmin = float(input("Enter the minimum time Tmin: "))
Tmax = float(input("Enter the maximum simulation time Tmax: "))
simulate_van_der_pol(m, Tmin, Tmax)
```

```python
#SIMPLIFIED VAN DER POL OSCILLATOR
#
#  X' = V
#  V' = M ( 1 - |X|)V - X  


import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the Van der Pol oscillator equations
def van_der_pol_simp(t, y, m):
    x, v = y
    dxdt = v
    dvdt = m * (1 - np.abs(x) ) * v - x
    return [dxdt, dvdt]

# Simulation function
def simulate_van_der_pol_simp(m, Tmin, Tmax):
    # Set random initial conditions in the rectangle [-0.1, 0.1] * [-0.1, 0.1]
    x0 = np.random.uniform(-0.1, 0.1)
    v0 = np.random.uniform(-0.1, 0.1)
    
    # Define the time span and evaluation points with a minimum of 30,000 points
    t_span = (0, Tmax)
    t_eval = np.linspace(Tmin, Tmax, 30000)  # Ensures 30,000 points in [Tmin, Tmax]
    
    # Solve the ODE using a stiff solver
    sol = solve_ivp(van_der_pol_simp, t_span, [x0, v0], args=(m,), t_eval=t_eval, method='LSODA')
    
    # Plot results in the interval [Tmin, Tmax]
    plt.figure(figsize=(10, 5))
    plt.plot(sol.t, sol.y[0], label='x (position)')
    plt.plot(sol.t, sol.y[1], label='v (velocity)')
    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.title(f'Van der Pol Oscillator Simulation (m={m}, Tmin={Tmin}, Tmax={Tmax})')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Phase plot
    plt.figure(figsize=(6, 6))
    plt.plot(sol.y[0], sol.y[1], color='purple')
    plt.xlabel('x (position)')
    plt.ylabel('v (velocity)')
    plt.title('Phase Plot of Van der Pol Oscillator')
    plt.grid()
    plt.show()

```

```python
# User input for m, Tmin, and Tmax
m = float(input("Enter the parameter m: "))
Tmin = float(input("Enter the minimum time Tmin: "))
Tmax = float(input("Enter the maximum simulation time Tmax: "))
simulate_van_der_pol_simp(m, Tmin, Tmax)
```
