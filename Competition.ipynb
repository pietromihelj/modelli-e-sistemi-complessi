```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the system of ODEs
def system(t, y, a, b, c):
    U, V = y
    dU_dt = U * (1 - U - a * V)
    dV_dt = c * V * (1 - b * U - V)
    return [dU_dt, dV_dt]

# Function to simulate the system
def simulate_system(a, b, c, U0, V0, T):
    # Time span
    t_span = (0, T)
    t_eval = np.linspace(0, T, 1000)
    
    # Initial conditions
    y0 = [U0, V0]
    
    # Solve the system of ODEs
    sol = solve_ivp(system, t_span, y0, args=(a, b, c), t_eval=t_eval)
    
    # Extract solutions
    U = sol.y[0]
    V = sol.y[1]
    t = sol.t
    
    # Plot U(t) and V(t)
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(t, U, label='U(t)', color='blue')
    plt.plot(t, V, label='V(t)', color='orange')
    plt.xlabel('t')
    plt.ylabel('U, V')
    plt.title('U(t) and V(t) vs Time')
    plt.legend()
    
    # Phase plane plot
    plt.subplot(1, 2, 2)
    plt.plot(U, V, color='green')
    plt.xlabel('U')
    plt.ylabel('V')
    plt.title('Phase Plane: U vs V')
    
    plt.tight_layout()
    plt.show()

# Parameters (replace these values with the ones you want to use)
a = 0.30  # Parameter a
b = 0.25  # Parameter b
c = 1.0  # Parameter c
T = 30.0  # Simulation time
U0 = 0.25  # Initial condition for U
V0 = 0.2  # Initial condition for V

# Run the simulation and plot
simulate_system(a, b, c, U0, V0, T)
```
