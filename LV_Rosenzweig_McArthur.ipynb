```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
c=2


eta = float(input("Enter a positive value for eta (>0): "))

rho = float(input("Enter a positive value for rho (>0): "))

Tmax = float(input("Enter a positive value for Tmax (>0): "))


# System of differential equations
def lotka_volterra(t, y):
    U, V = y
    dU_dt = U*(1-U)  -  U * V/(1+eta*U)
    dV_dt = c*V*( -1 + rho*U/(1+eta*U)  )
    return [dU_dt, dV_dt]

# Initial conditions
U0 = 0.1  # Initial prey population
V0 = 0.1  # Initial predator population
y0 = [U0, V0]

# Time span for the simulation
t_span = (0, Tmax)  
t_eval = np.linspace(t_span[0], t_span[1], 10000)  

# Solve the ODEs using solve_ivp
solution = solve_ivp(
    lotka_volterra, t_span, y0, t_eval=t_eval, method='RK45',
    rtol=1e-6, atol=1e-9
)

# Extract solutions
U = solution.y[0]
V = solution.y[1]
t = solution.t

# Plotting the results
plt.figure(figsize=(14, 6))

# Plot U(t) and V(t) over time
plt.subplot(1, 2, 1)
plt.plot(t, U, label='Prey Population U(t)', color='blue')
plt.plot(t, V, label='Predator Population V(t)', color='red')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Population Dynamics Over Time')
plt.legend()
plt.grid(True)

# Phase plane plot (V vs U)
plt.subplot(1, 2, 2)
plt.plot(U, V, color='green')
plt.xlabel('Prey Population U')
plt.ylabel('Predator Population V')
plt.title('Phase Plane Trajectory (V vs U)')
plt.grid(True)

plt.tight_layout()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
c=2




rho = float(input("Enter a positive value for rho (>0): "))

eta = float(input("Enter a positive value for eta (>0): "))

epsi = float(input("Enter a positive value for epsi (>0): "))



# System of differential equations
def lotka_volterra(t, y):
    U, V = y
    dU_dt = U*(1-U)  -  U * V/(1+eta*U)
    dV_dt = c*V*( -1 - epsi*V + rho*U/(1+eta*U)  )
    return [dU_dt, dV_dt]

# Initial conditions
U0 = 0.1  # Initial prey population
V0 = 0.1  # Initial predator population
y0 = [U0, V0]

# Time span for the simulation
t_span = (0, 200)  
t_eval = np.linspace(t_span[0], t_span[1], 10000)  

# Solve the ODEs using solve_ivp
solution = solve_ivp(
    lotka_volterra, t_span, y0, t_eval=t_eval, method='RK45',
    rtol=1e-6, atol=1e-9
)

# Extract solutions
U = solution.y[0]
V = solution.y[1]
t = solution.t

# Plotting the results
plt.figure(figsize=(14, 6))

# Plot U(t) and V(t) over time
plt.subplot(1, 2, 1)
plt.plot(t, U, label='Prey Population U(t)', color='blue')
plt.plot(t, V, label='Predator Population V(t)', color='red')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Population Dynamics Over Time')
plt.legend()
plt.grid(True)

# Phase plane plot (V vs U)
plt.subplot(1, 2, 2)
plt.plot(U, V, color='green')
plt.xlabel('Prey Population U')
plt.ylabel('Predator Population V')
plt.title('Phase Plane Trajectory (V vs U)')
plt.grid(True)

plt.tight_layout()
plt.show()

```
